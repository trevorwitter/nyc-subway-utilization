```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/mta_subway_220108_dbscan.csv')
df1 = pd.read_csv('../data/mta_subway_220108.csv')
df = df.set_index('TIME')
df1 = df1.set_index('TIME')
```

```python
df.max().max()
```

```python
df1.max().max()
```

```python
def plot_ents(station):
    df[f'{station}_ent'].plot()
    df[f'{station}_ex'].plot()
    df1[f'{station}_ent'].plot(alpha=0.3)
    df1[f'{station}_ex'].plot(alpha=0.3)
    plt.show()
```

```python
plot_ents('103 ST')
```
