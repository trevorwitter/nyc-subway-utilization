```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_loc = "../../data/mta_subway_221231_100wk_dbscan.parquet"
```

```python
df = pd.read_parquet(file_loc)
df.shape
```

```python
stations = set([x.split("_")[0] for x in df.columns])
```

```python
def plot_ents(station):
    df[f'{station}_ent'].plot(alpha=0.3)
    df[f'{station}_ex'].plot(alpha=0.3)
    plt.show()
```

```python
plot_ents('14 ST')
```

```python
plot_ents('GRD CNTRL-42 ST')
```

```python
plot_ents('14 ST-UNION SQ')
```

```python
plot_ents('7 AV')
```
